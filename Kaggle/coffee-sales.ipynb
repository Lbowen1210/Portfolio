# Bar Chart

```python
# Sample data

categories = df_coffee['cash_type']

values = df_coffee['money']

# Create a bar plot
plt.figure(figsize=(4,4))
plt.bar(categories, values,color='skyblue',width = 0.5)
plt.title('Amount generated per coffee variant')
plt.xlabel('Coffee Name',fontweight='bold', color='orange', fontsize=10)
plt.xticks(rotation=45)
plt.ylabel('Amount',fontweight='bold', color='green', fontsize=10)
plt.legend()
plt.show()
```

# Scatter plot

```python
# scatter plot
plt.figure(figsize=(4,6))
sns.scatterplot(data=df_coffee, x='cash_type', y='money', hue='coffee_name',palette='coolwarm')
plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
sns.set_theme(style="dark")
plt.figure(figsize=(4,6))
sns.relplot(data=df_coffee, x='cash_type', y='coffee_name', size='money',hue='money',palette='coolwarm')
#plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
